# Categorize Drugs

```python
illegal_drugs2 = ['Meth', 'Benzos', 'Heroin', 'Crack']
for drug in illegal_drugs2:
    df[drug+'_bin'] = [1 if d > 2 else 0 for d in df[drug].values]
df['illegal_drugs_2'] = df['Meth_bin'] + df['Benzos_bin'] + df['Heroin_bin'] + df['Crack_bin']
df['illegal_drugs_2'] = [1 if ill > 0 else 0 for ill in df.illegal_drugs_2.values]
df.illegal_drugs_2.value_counts() 
```

```python
df['Alcohol_bin'] = [1 if alc >= 5 else 0 for alc in df.Alcohol.values]
df.Alcohol_bin.value_counts()
```

```python
df['Nicotine_bin'] = [1 if nic > 5 else 0 for nic in df.Nicotine.values]
```

```python
illegal_drugs1 = ['Cannabis', 'Legalh', 'LSD', 'Shrooms', 
                  'Amphet', 'Amyl', 'Ecstasy', 'Ketamine', 
                  'Coke']


borders = [4, 2, 3, 3, 3, 3, 3, 2, 2]
drug_zip = list(zip(borders, illegal_drugs1))
for b, drug in zip(borders, illegal_drugs1):
    df[drug+'_bin'] = [1 if d > b else 0 for d in df[drug].values]
```

```python
illegal_drugs1_bins = []
for drug in illegal_drugs1:
    illegal_drugs1_bins.append(drug+"_bin")
```

```python
df['illegal_drugs_1'] = df[illegal_drugs1_bins].sum(axis=1)
```

```python
df['illegal_drugs_1'] = [1 if ill > 0 else 0 for ill in df.illegal_drugs_1.values]
df.illegal_drugs_1.value_counts()
```

```python
df
```

# Age distribution for each Drug category

```python
drug_cats = ['Alcohol_bin', 'Nicotine_bin', 'illegal_drugs_1', 'illegal_drugs_2']
drug_titles = ['Alcohol', 'Nicotine', '"Party" Drugs', '"Junkie" Drugs']
```

```python
age_grouped_mean = df.groupby("Age").mean()
age_grouped_mean
```

```python
fig = plt.figure(figsize=(8,8))
for i, drug in enumerate(drug_cats):
    sns.lineplot(x=age_grouped_mean.index, y=age_grouped_mean[drug], marker='s', linewidth=3, markersize=8,
             data=age_grouped_mean, label=drug_titles[i])

plt.xlabel("Age", fontsize=18, labelpad=10)  
plt.ylabel("Mean User", fontsize=18, labelpad=10) 
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=14)
plt.grid(True)
```

# Education versus Drug Categories

```python
edu_grouped_mean = df.groupby("Education").mean()
edu_grouped_mean
```

```python
fig = plt.figure(figsize=(8,8))
for i, drug in enumerate(drug_cats):
    sns.lineplot(x=edu_grouped_mean.index, y=edu_grouped_mean[drug], marker='s', linewidth=3, markersize=8,
             data=edu_grouped_mean, label=drug_titles[i])

plt.xlabel("Education", fontsize=18, labelpad=10)  
plt.ylabel("Mean User", fontsize=18, labelpad=10) 
plt.xticks(fontsize=16, rotation=45)
plt.yticks(fontsize=16)
plt.legend(fontsize=14)
plt.grid(True)
```

## Age Distribution

```python
age_sorted = df.sort_values("Age")
age_grouped_mean = age_sorted.groupby("Age").mean()
age_grouped_median = age_sorted.groupby("Age").median()
age_grouped_mean
age_grouped_median
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

sns.histplot(x='Age',data=age_sorted,color='blue',ax=ax1)
sns.lineplot(x=age_grouped_mean.index, y='Gender', data=age_grouped_mean,
             marker='s',color='orange',ax=ax2, linewidth=3)
ax1.set_xlabel('Age', fontsize=16)
ax1.set_ylabel('Counts', fontsize=16, labelpad=10)
ax2.set_ylabel('Gender', fontsize=16, labelpad=10)
ax1.tick_params(labelsize=16)
ax2.tick_params(labelsize=16)

plt.show()
```
