```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

pd.set_option("display.max_columns", None)
sns.set_style("white")
```

```python
df = pd.read_csv("../../data/processed/data_test_eda.csv")
```

```python
df.head()
```

```python
df.describe()
```

## Age Distribution

```python
age_sorted = df.sort_values("Age")
age_grouped_mean = age_sorted.groupby("Age").mean()
age_grouped_median = age_sorted.groupby("Age").median()
age_grouped_mean
age_grouped_median
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

sns.histplot(x='Age',data=age_sorted,color='blue',ax=ax1)
sns.lineplot(x=age_grouped_mean.index, y='Gender', data=age_grouped_mean,
             marker='s',color='orange',ax=ax2, linewidth=3)
ax1.set_xlabel('Age', fontsize=16)
ax1.set_ylabel('Counts', fontsize=16, labelpad=10)
ax2.set_ylabel('Gender', fontsize=16, labelpad=10)
ax1.tick_params(labelsize=16)
ax2.tick_params(labelsize=16)

plt.show()
```

```python
fig_sunburst = px.sunburst(df, path=['Age', 'Country'])
fig_sunburst.show()
```

```python
markersize = 10
linewidth = 3

fig, ax = plt.subplots(figsize=(10,10))

sns.lineplot(x=age_grouped_median.index, y='Nscore', data=age_grouped_median,
             marker='s',color='orange', label="Neuroticism", linewidth=linewidth,
             markersize=markersize)
sns.lineplot(x=age_grouped_median.index, y='Escore', data=age_grouped_median,
             marker='o',color='blue', label="Extraversion")
sns.lineplot(x=age_grouped_median.index, y='Oscore', data=age_grouped_median,
             marker='o',color='red', label="Openness", linewidth=linewidth,
             markersize=markersize)
sns.lineplot(x=age_grouped_median.index, y='Ascore', data=age_grouped_median,
             marker='o',color='black', label="Agreeableness")
sns.lineplot(x=age_grouped_median.index, y='Cscore', data=age_grouped_median,
             marker='o',color='green', label="Conscientiousness", linewidth=linewidth,
             markersize=markersize)

ax.set_xlabel('Age', fontsize=16)
ax.set_ylabel('Mean Personality Scores', fontsize=16, labelpad=10)

ax.tick_params(labelsize=16)
plt.legend(fontsize=16)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

sns.lineplot(x=age_grouped_median.index, y='Impulsive', data=age_grouped_median,
             marker='s',color='orange', label="Impulsiveness", linewidth=linewidth,
             markersize=markersize)
sns.lineplot(x=age_grouped_median.index, y='SS', data=age_grouped_median,
             marker='o',color='blue', label="Sensation seeing", linewidth=linewidth,
             markersize=markersize)

ax.set_xlabel('Age', fontsize=16)
ax.set_ylabel('Mean Personality Scores', fontsize=16, labelpad=10)

ax.tick_params(labelsize=16)
plt.legend(fontsize=16)
```

# Drug Use per age

```python
df.head()
```

```python
age_grouped_sum = df.groupby("Age").sum()
age_grouped_sum
drugs = age_grouped_sum.iloc[0,8:-18].index.tolist()
drugs1 = drugs[:6]
drugs2 = drugs[6:12]
drugs3 = drugs[12:]
```

```python
fig = plt.figure(figsize=(12,8))
for drug in drugs1:
    sns.lineplot(x=age_grouped_sum.index, y=age_grouped_sum[drug]/df.Age.value_counts(),
             data=age_grouped_sum, label=drug)
```

```python
fig = plt.figure(figsize=(12,8))
for drug in drugs2:
    sns.lineplot(x=age_grouped_sum.index, y=age_grouped_sum[drug]/df.Age.value_counts(),
             data=age_grouped_sum, label=drug)
```

```python
fig = plt.figure(figsize=(12,8))
for drug in drugs3:
    sns.lineplot(x=age_grouped_sum.index, y=age_grouped_sum[drug]/df.Age.value_counts(),
             data=age_grouped_sum, label=drug)
```
